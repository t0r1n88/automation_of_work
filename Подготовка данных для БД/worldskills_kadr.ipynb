```python
import pandas as pd
import datetime
```

```python
# Считываем данные
dm = pd.read_excel('../data/DM.xlsx',)
rch = pd.read_excel('../data/RCH.xlsx')
```

```python
# Конвертируем столбец дата выдачи из int в datetime
dm['Дата выдачи'] = pd.to_datetime(dm['Дата выдачи'])
rch['Дата выдачи'] = pd.to_datetime(rch['Дата выдачи'])
```

```python
# Создаем переменную для отсчета от сегодняшнего дня
today = datetime.datetime.today()
# Длительность действия удостоверения
duration = 2
```

```python
# Создаем условие для отбора
duration_condition = today - datetime.timedelta(days =730)
```

```python
# Создаем 2 новых 2 датасета, чтобы не путаться
time_condition_dm  = dm[dm['Дата выдачи'] > duration_condition]
time_condition_rch  = rch[rch['Дата выдачи'] > duration_condition]
```

```python
# Создаем новый столбец где будет хранится уровень полномочий эксперта
time_condition_dm['Уровень полномочий'] = 'ДЭ'
time_condition_rch['Уровень полномочий'] = 'РЧ'
```

```python
# Объединение датасетов
df = time_condition_rch.append(time_condition_dm)
```
